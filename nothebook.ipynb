```python
#import library
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#load datset
df = pd.read_csv("Books_Data_Clean.csv")
```

```python
#display dataset
df.head()
```

```python
#check the numerical summary
df.describe()
```

```python
#fix publishng year column  - filter
df = df[df["Publishing Year"] > 1900]
```

```python
#check missing values
df.isna().sum()
```

```python
# remove na in book name column
df.dropna(subset="Book Name", inplace= True)
```

```python
#check for duplicates
df.duplicated().sum()
```

```python
#to check the unique values in a column
df.nunique()
```

```python
#histogrm distribution of publishing year
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution of Publishing year")
plt.show()
```

```python
#distribution of genre
df["genre"].value_counts().plot(kind="bar")
plt.xlabel("Genre")
plt.ylabel("Frequency")
plt.title("Distribution of Genre")
plt.show()
```

```python
#groupby author and book rating
df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending= False)
```

```python
df.nunique()
```

```python
sns.boxplot(x ="genre", y = "Book_ratings_count", data= df)
plt.xlabel("Genre")
plt.ylabel("Book Rating Count")
plt.title("Number of books by each Genre")
plt.show()
```

```python
#scatter plot sales price by unit sold
df.nunique()
plt.scatter(df["sale price"], df["units sold"])
plt.xlabel("Sales Price")
plt.ylabel("Unit Sold")
plt.title("Scatter plot sales price vs unit Sold")
plt.show()
```

```python
#value count of language code
language_count = df["language_code"].value_counts()
```

```python
#plot a pie chart of language distribution
plt.pie(language_count, labels = language_count.index, startangle= 90)
plt.title("Language Distribution")
```

```python
#group by published year by publisher
df.groupby("Publisher ")["publisher revenue"].sum().sort_values(ascending= False)
```

```python
df.columns
```

```python
#group by author vs book rating - max
df.groupby("Author_Rating")["Book_ratings_count"].mean().sort_values(ascending= False).max()
```

```python
#count of language code
df.groupby("language_code").size().sort_values(ascending= False)
```

```python
#group by author vs book rating
df.groupby("Author_Rating")["Book_ratings_count"].max()
```

```python
#scatter plot average rating vs rating count
plt.scatter(df["Book_average_rating"], df["Book_ratings_count"])
plt.xlabel("Average Rating")
plt.ylabel("Rating Count")
plt.title("Scatter plot of Average Rating vs Rating Count")
plt.show()
```

```python
#group by author by gross sales
total_gross_sale_by_author = df.groupby("Author")["gross sales"].sum()
total_gross_sale_by_author.sort_values(ascending= False).head(10)
```

```python
#bar chart total gross sales by author
total_gross_sale_by_author.sort_values(ascending= False).head(10).plot(kind="bar")
plt.xlabel("Author")
plt.ylabel("Gross Sales")
plt.title("Total gross sales by author")
plt.show()
```

```python
#box plot unit sold vs autuhor rating
df.nunique()
sns.boxplot(x = "Author_Rating", y = "units sold", data= df)
plt.ylabel("Units Sold")
plt.xlabel("Author Rating")
plt.title("Boxplot of Unit Sold vs Author Rating")
plt.show()
```

```python
#line chart publishing year vs unit sold
df.groupby("Publishing Year")["units sold"].sum().plot(kind="line", marker = "x")
plt.ylabel("Unit Sold")
plt.xlabel("Publishing Year")
plt.title("Total Unit Sold over the Years")
plt.show()
```
